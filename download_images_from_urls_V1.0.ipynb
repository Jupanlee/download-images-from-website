```python
import os
import requests
from bs4 import BeautifulSoup

site_url = 'https://shopdoen.com/pages/spring-18'

# http GET 请求

# 超文本传输协议（Hyper Text Transmission Protocol, HTTP）用于保证客户端与服务器之间的通信，
# 工作方式为客户机与服务器之间的请求-应答协议。客户端向服务器发出HTTP请求，服务器向客户端返
# 回响应。

# GET 和 POST 是两种最常用的方法。

# GET：从指定的资源请求数据。查询字符串（名称/值对）是在GET请求的URL中发送的。例：/test/demo_form.asp?name1=value1&name2=value2
# 1. GET 请求可被缓存
# 2. GET 请求保留在浏览器历史记录中
# 3. GET 请求可被收藏为书签
# 4. GET 请求不应在处理敏感数据时使用
# 5. GET 请求有长度限制
# 6. GET 请求只应当用于取回数据

# POST：向指定的资源提交要被处理的数据。查询字符串（名称/值对）是在 POST 请求的 HTTP 消息主体中发送的
# 1. POST 请求不会被缓存
# 2. POST 请求不会保留在浏览器历史记录中
# 3. POST 不能被收藏为书签
# 4. POST 请求对数据长度没有要求

# 参考：http://www.w3school.com.cn/tags/html_ref_httpmethods.asp

r = requests.get(site_url)
```

```python
# BeautifulSoup 主要功能：从网页抓取数据，支持Python标准库中的HTML解析器
soup = BeautifulSoup(r.text,'lxml')
print(soup)
images = soup.findAll("img")
print("Found " + str(len(images)) + " images.")

for image in images:
    print(image)
    try:
        filename = image["src"].split("/")[-1].split("?")[0]
        print(filename)
       
        # rstrip() 用于删除字符串末尾的指定字符；isalpha()判断字符是否为英文字母      
        # filename = "".join([c for c in filename if c.isalpha() or c.isdigit() or c==' ' or c=='.']).rstrip() # Only valid OS chars
        filename = os.path.join(os.getcwd(), filename)
        
        # Fix relative images
        if (not image["src"].lower().startswith("http") and not image["src"].lower().startswith("data") ):
            # Relative image add full path
            image["src"] = "https:" + image["src"]
            print(image["src"])
            print("ok")
        print('ok')
        print(image["src"])
        response = requests.get(image["src"])
        print(response)
        if response.status_code == 200:
            f = open(filename, 'wb')
            f.write(response.content)
            f.close()


    except Exception as e:
        traceback.print_exc()
        print(str(e))
```
